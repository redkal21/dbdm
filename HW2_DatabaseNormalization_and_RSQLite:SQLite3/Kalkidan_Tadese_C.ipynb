```python
#install sqlite3 and pandas
import sqlite3
import pandas as pd
```

```python
#connect to sqlite database
con = sqlite3.connect("Kalkidan_Tadese_db.sqlite")
c = con.cursor() 
# create pointer/cursor object to interact with db
```

```python
# query sqlite_master table to retrieve db metadata
Dtables = pd.read_sql("""SELECT *
FROM sqlite_master
WHERE type='table';""", con)
Dtables #display metadata 
```

```python
data_tbl = pd.read_sql("""SELECT * FROM Data;""",con)
# query data table, select everything
data_tbl.dtypes # display the datatypes of columns in Data table
```

```python
avg_exp = data_tbl.groupby(["probeid", "expid"])["expvalue"].mean()
#group data by probeid and expid, then calculate mean of expvalue
print(avg_exp.head())
#print last few rows
```
